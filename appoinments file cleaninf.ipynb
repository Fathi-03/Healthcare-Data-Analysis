```python
# make temp column for fees column
temp_fees = df['fees'].copy()
```

```python
# replace null values in column discount with mode value
df['discount']=df['discount'].fillna(df['discount'].mode()[0])
```

```python
# replace null values in column payment_mathod with mode value
df['payment_method']=df['payment_method'].replace('0',df['payment_method'].mode()[0])
```
